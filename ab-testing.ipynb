```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
control_df = pd.read_csv("hackaton_control_group.csv")
control_df.head(10)
```

```python
test_df = pd.read_csv("hackaton_test_group.csv")
test_df.head(10)
```

```python
control_df["CTR"] = control_df["# of Website Clicks"] / control_df["# of Impressions"]
control_df["Click Cost"] = control_df["# of Website Clicks"] / control_df["Spend [USD]"]
control_df.head(10)
```

```python
test_df["CTR"] = test_df["# of Website Clicks"] / test_df["# of Impressions"]
test_df["Click Cost"] = test_df["# of Website Clicks"] / test_df["Spend [USD]"]
test_df.head(10)
```

```python
a1 = control_df["# of Purchase"] / control_df["# of Website Clicks"]
a2 = test_df["# of Purchase"] / test_df["# of Website Clicks"]
plt.hist(a1, alpha = 0.5)
plt.hist(a2, alpha = 0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
# blue is control, orange is test group
```

```python
a1 = control_df["# of Website Clicks"] / control_df["# of Impressions"] / control_df["Spend [USD]"]
a2 = test_df["# of Website Clicks"] / test_df["# of Impressions"] / test_df["Spend [USD]"]
plt.plot(a1, label='Control Group')
plt.plot(a2, label='Test Group')
plt.legend()
```

```python
a1 = control_df["# of Purchase"] / control_df["Spend [USD]"]
a2 = test_df["# of Purchase"] / test_df["Spend [USD]"]
plt.plot(a1, label='Control Group')
plt.plot(a2, label='Test Group')
plt.legend()
```

```python
a1 = control_df["# of Purchase"] / control_df["# of Impressions"]
a2 = test_df["# of Purchase"] / test_df["# of Impressions"]
plt.plot(a1, label='Control Group')
plt.plot(a2, label='Test Group')
plt.legend()
```

```python
a1 = control_df["# of Impressions"] / control_df["Spend [USD]"]
a2 = test_df["# of Impressions"] / test_df["Spend [USD]"]
plt.plot(a1, label='Control Group')
plt.plot(a2, label='Test Group')
plt.legend()
```

```python
a1 = control_df["# of Purchase"] / control_df["Spend [USD]"]
a2 = test_df["# of Purchase"] / test_df["Spend [USD]"]
plt.scatter(control_df["Date"], a1, alpha = 0.5)
plt.scatter(test_df["Date"], a2, alpha = 0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
# blue is control, orange is test group
```
